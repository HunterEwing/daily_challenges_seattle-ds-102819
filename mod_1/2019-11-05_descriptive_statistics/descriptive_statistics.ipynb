**1) What are the mean age and the median age for the players in this dataset?** 

The column is named 'Age'

```python
# code here
print(df["Age"].mean())
print(df["Age"].median())
df.shape
```

In your own words, how are the mean and median related to each other and what do these values tell us about the distribution of the players' ages?

```python
"""
The way that the mean and median diverge relative to a perfect normal 
distribution gives insight into skewness. If we have reason to believe that our 
distribution is roughly Gaussian, then if our distribution's mean is to the 
right of (i.e., higher than) the median, the distribution is 
right-skwewed - meaning we have high-value outliers. If the mean is lower 
than the median, then we have a left-skewed distribution. 

The fact that the mean is slightly above 
the median means that the distribution 
of player ages is slightly right-skewed."""
```
